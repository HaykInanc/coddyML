## Теоритическая часть

Програмный продукт, использующий технологию машинного обучения, представляет из себя приложение, которое опираясь на историю событий в прошлом может делать предсказания. 

Первым делом необходимо сформировать множество данных, на которых машина "обучается". 
Эти данные должны иметь вид [множество параметров] -> ответ. 

Далее программа, используя один из алгоритмов принятия решений, способна выдавать ответ, получая на вход [множество параметров]. 

## Алгоритм KNN

Данный алгоритм строит множество точек в пространстве, используя [множество параметров], как координаты. Колличество параметров соответствует колличеству измерений (человек может представить только три измерения, но этого ограничения нет у компьютера). 

При фрмировании "предсказания" машина определяет K ближайших соседий и определяет их свойства. Какое свойство превосходит в колличестве, то и выводится, как ответ. 

## Граффическое пояснение 



предположим, что x1 и x2 это свойства параметров, а class A и class B возможные ответы машины. 
В зависимости от K рассматривается соответствующее кол-во соседей и присуждается тот Class, который чаще встречается среди выбранных эллементов.

## Практическая часть 

Ниже представлен пример, который мы рассматривали на уроке.

```python
from sklearn.datasets import load_iris                 #подключение массива данных для примера. Массив с цветочками Iris
from sklearn.neighbors import KNeighborsClassifier     #подключение реализации алгоритма KNN

iris = load_iris()                                     #загрузка данных по цветкам

X = iris.data                                          #массив с параметрами цветков
y = iris.target                                        #массив с подтипами цветков

knn = KNeighborsClassifier(n_neighbors=1)              #формирование машины с алгоритмом KNN, где число соседей = 1

knn.fit(X, y)                                          #процес обучения машины, на вход подаются значения параметров и ответы

print(knn.predict([[3, 5, 0, 2]]))                     #запуск "предсказания" и вывод в консоль типа цветка
```
