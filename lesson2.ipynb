```python
from sklearn.datasets import load_iris                 #    .    Iris
from sklearn.neighbors import KNeighborsClassifier     #   KNN

iris = load_iris()                                     #   

X = iris.data                                          #   
y = iris.target                                        #   

knn = KNeighborsClassifier(n_neighbors=1)              #    KNN,    = 1

knn.fit(X, y)                                          #  ,       

print(knn.predict([[3, 5, 0, 2]]))                     # ""      
```
